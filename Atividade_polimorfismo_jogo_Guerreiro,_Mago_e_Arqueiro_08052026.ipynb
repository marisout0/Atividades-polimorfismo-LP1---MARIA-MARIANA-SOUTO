```python
%%writefile jogoGMA.cpp
#include <iostream>
#include <string>
#include <vector>
#include <cmath>

using namespace std;

// Classe Base
class Personagem {
protected:
    string nome;
    int hp, hpMax;
    int mp, mpMax;

public:
    Personagem(string n, int h, int m) : nome(n), hp(h), hpMax(h), mp(m), mpMax(m) {}

    virtual void ataqueSimples(Personagem &alvo) = 0; // Método virtual
    virtual void ataqueEspecial(Personagem &alvo) = 0;

    void recuperarTurno() {
        // Recupera 10% de HP e MP máximo
        hp = min(hpMax, hp + (int)(hpMax * 0.1));
        mp = min(mpMax, mp + (int)(mpMax * 0.1));
        cout << nome << " recuperou HP e MP!" << endl;
    }

    bool estaVivo() { return hp > 0; } // Perde quem chegar a <= 0 HP

    void receberDano(int dano) {
        hp -= dano;
        cout << nome << " recebeu " << dano << " de dano. HP atual: " << hp << endl;
    }

    string getNome() { return nome; }
    int getMP() { return mp; }
    void consumirMP(int custo) { mp -= custo; }
};

// Classes Filhas
class Guerreiro : public Personagem {
public:
    Guerreiro(string n) : Personagem(n, 150, 50) {}

    void ataqueSimples(Personagem &alvo) override {
        cout << nome << " ataca com a espada!" << endl;
        alvo.receberDano(20);
    }

    void ataqueEspecial(Personagem &alvo) override {
        if (getMP() >= (int)(50 * 0.15)) { // Consome 15% do MP máximo
            cout << nome << " usa CORTE DIMENSIONAL!" << endl;
            consumirMP(7);
            alvo.receberDano(45);
        } else {
            cout << "MP insuficiente para ataque especial!" << endl;
        }
    }
};

class Mago : public Personagem {
public:
    Mago(string n) : Personagem(n, 80, 150) {}

    void ataqueSimples(Personagem &alvo) override {
        cout << nome << " lança magia!" << endl;
        alvo.receberDano(15);
    }

    void ataqueEspecial(Personagem &alvo) override {
        if (getMP() >= (int)(150 * 0.15)) {
            cout << nome << " lança MAGIA SUPREMA!" << endl;
            consumirMP(22);
            alvo.receberDano(60);
        } else {
            cout << "MP insuficiente!" << endl;
        }
    }
};

class Arqueiro : public Personagem {
public:
    Arqueiro(string n) : Personagem(n, 100, 80) {}

    void ataqueSimples(Personagem &alvo) override {
        cout << nome << " lança flecha!" << endl;
        alvo.receberDano(18);
    }

    void ataqueEspecial(Personagem &alvo) override {
        if (getMP() >= (int)(80 * 0.15)) {
            cout << nome << " lança FLECHA PERFURANTE!" << endl;
            consumirMP(12);
            alvo.receberDano(40);
        } else {
            cout << "MP insuficiente!" << endl;
        }
    }
};

int main() {
    Guerreiro g("Thorin");
    Mago m("Gandalf");

    // Simulação de turno escolhido pelo usuário
    while (g.estaVivo() && m.estaVivo()) {
        g.recuperarTurno();
        int op;
        cout << "\nTurno de " << g.getNome() << " (1-Simples, 2-Especial): ";
        cin >> op;

        if (op == 1) g.ataqueSimples(m);
        else g.ataqueEspecial(m);

        if (!m.estaVivo()) break;

        m.recuperarTurno();
        cout << "\nTurno de " << m.getNome() << " (1-Simples, 2-Especial): ";
        cin >> op;

        if (op == 1) m.ataqueSimples(g);
        else m.ataqueEspecial(g);
    }

    cout << "\nFim de jogo!" << endl;
    return 0;
}
```
