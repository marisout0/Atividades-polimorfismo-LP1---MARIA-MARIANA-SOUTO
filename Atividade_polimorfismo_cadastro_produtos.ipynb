```python
%%writefile cadastroprodutos.cpp
// writefile é um comando usado para que eu possa programar em outras linguagens no colab
#include <iostream>
#include <vector>
#include <string>
#include <limits> // Necessário para numeric_limits

using namespace std;

// Enum para o tipo de uso dos produtos de limpeza
enum TipoUso { DOMESTICO, HOSPITALAR, INDUSTRIAL };

// Classe Base: Produto
class Produto {
protected:
    string nome;
    double preco;

public:
    Produto(string n, double p) : nome(n), preco(p) {}

    // Método virtual para permitir polimorfismo na exibição
    virtual void exibirInformacoes() {
        cout << "Nome: " << nome << " | Preco: R$ " << preco;
    }

    // Destrutor virtual para liberar a memória das subclasses corretamente
    virtual ~Produto() {}
};

// Subclasse 1: Produto de Limpeza
class ProdutoLimpeza : public Produto {
    TipoUso tipo;
public:
    ProdutoLimpeza(string n, double p, TipoUso t) : Produto(n, p), tipo(t) {}

    void exibirInformacoes() override {
        Produto::exibirInformacoes(); // Reutiliza código da base
        string tStr;
        if(tipo == DOMESTICO) tStr = "Domestico";
        else if(tipo == HOSPITALAR) tStr = "HOSPITALAR";
        else tStr = "Industrial";
        cout << " | Tipo: " << tStr << " (Limpeza)" << endl;
    }
};

//Subclasse 2: Produto Alimenticio
class ProdutoAlimenticio : public Produto {
    string validade;
public:
    ProdutoAlimenticio(string n, double p, string v) : Produto(n, p), validade(v) {}

    void exibirInformacoes() override {
        Produto::exibirInformacoes();
        cout << " | Validade: " << validade << " (Alimento)" << endl;
    }
};

// Subclasse 3: Produto Eletronico
class ProdutoEletronico : public Produto {
    int voltagem;
    int garantia;
public:
    ProdutoEletronico(string n, double p, int v, int g) : Produto(n, p), voltagem(v), garantia(g) {}

    void exibirInformacoes() override {
        Produto::exibirInformacoes();
        cout << " | Voltagem: " << voltagem << "V | Garantia: " << garantia << " meses (Eletronico)" << endl;
    }
};

// Função de entrada para inteiros
int getIntInput(const string& prompt) {
    int value;
    while (true) {
        cout << prompt;
        cin >> value;
        if (cin.fail()) {
            cout << "Entrada inválida. Por favor, digite um número inteiro." << endl;
            cin.clear(); // Limpa os sinalizadores de erro
            cin.ignore(numeric_limits<streamsize>::max(), '\n'); // Descarta a entrada inválida
        } else {
            cin.ignore(numeric_limits<streamsize>::max(), '\n'); // Descarta o restante da linha (nova linha)
            return value;
        }
    }
}

// Função de entrada para doubles
double getDoubleInput(const string& prompt) {
    double value;
    while (true) {
        cout << prompt;
        cin >> value;
        if (cin.fail()) {
            cout << "Entrada inválida. Por favor, digite um número decimal." << endl;
            cin.clear(); // Limpa os sinalizadores de erro
            cin.ignore(numeric_limits<streamsize>::max(), '\n'); // Descarta a entrada inválida
        } else {
            cin.ignore(numeric_limits<streamsize>::max(), '\n'); // Descarta o restante da linha (nova linha)
            return value;
        }
    }
}

// Função de entrada para strings após uma entrada numérica
string getLineInput(const string& prompt) {
    string value;
    cout << prompt;
    getline(cin, value);
    return value;
}


int main() {
    vector<Produto*> listaProdutos;
    int choice;

    do {
        cout << "\n--- MENU DE CADASTRO DE PRODUTOS ---" << endl;
        cout << "1. Cadastrar Produto de Limpeza" << endl;
        cout << "2. Cadastrar Produto Alimentício" << endl;
        cout << "3. Cadastrar Produto Eletrônico" << endl;
        cout << "4. Exibir Produtos Cadastrados" << endl;
        cout << "0. Sair" << endl;
        choice = getIntInput("Escolha uma opção: ");

        switch (choice) {
            case 1: { // Produto de Limpeza
                string nome = getLineInput("Nome do produto de limpeza: ");
                double preco = getDoubleInput("Preço: R$");
                int tipoUsoInt = getIntInput("Tipo de uso (0: DOMÉSTICO, 1: HOSPITALAR, 2: INDUSTRIAL): ");
                TipoUso tipoUso = static_cast<TipoUso>(tipoUsoInt);
                listaProdutos.push_back(new ProdutoLimpeza(nome, preco, tipoUso));
                cout << "Produto de limpeza cadastrado com sucesso!" << endl;
                break;
            }
            case 2: { // Produto Alimenticio
                string nome = getLineInput("Nome do produto alimentício: ");
                double preco = getDoubleInput("Preço: R$");
                string validade = getLineInput("Validade (ex: 12/2025): ");
                listaProdutos.push_back(new ProdutoAlimenticio(nome, preco, validade));
                cout << "Produto alimentício cadastrado com sucesso!" << endl;
                break;
            }
            case 3: { // Produto Eletronico
                string nome = getLineInput("Nome do produto eletrônico: ");
                double preco = getDoubleInput("Preço: R$");
                int voltagem = getIntInput("Voltagem (V): ");
                int garantia = getIntInput("Garantia (meses): ");
                listaProdutos.push_back(new ProdutoEletronico(nome, preco, voltagem, garantia));
                cout << "Produto eletrônico cadastrado com sucesso!" << endl;
                break;
            }
            case 4: {
                cout << "\n--- LISTA DE PRODUTOS CADASTRADOS ---" << endl;
                if (listaProdutos.empty()) {
                    cout << "Nenhum produto cadastrado ainda." << endl;
                } else {
                    for (auto p : listaProdutos) {
                        p->exibirInformacoes();
                    }
                }
                break;
            }
            case 0:
                cout << "Saindo do programa." << endl;
                break;
            default:
                cout << "Opção inválida. Tente novamente." << endl;
                break;
        }
    } while (choice != 0);

    // Limpeza de memória
    for (auto p : listaProdutos) {
        delete p;
    }
    listaProdutos.clear();

    return 0;
}
```
