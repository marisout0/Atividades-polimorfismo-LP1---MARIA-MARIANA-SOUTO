```python
%%writefile autenticacaodiscord.cpp
#include <iostream>
#include <string>
#include <vector>

using namespace std;

// Classe Base Abstrata
class Autenticacao {
public:
    // Metodo principal: coletar -> validar -> iniciar
    void autenticar() {
        coletarDados(); // Chama a versao da subclasse
        if (validarDados()) {
            iniciarSessao();
        } else {
            cout << "Erro: Dados invalidos. Autenticacao falhou!" << endl;
        }
    }

    // Metodos abstratos (virtuais puros) que as filhas SAO OBRIGADAS a fazer
    virtual void coletarDados() = 0;
    virtual bool validarDados() = 0;

    // Metodo comum para todas as formas de login
    void iniciarSessao() {
        cout << "Sucesso! Iniciando sessao no Discord... Bem-vindo!" << endl;
    }

    virtual ~Autenticacao() {} // Destrutor virtual por seguranca
};

// Forma 1: Login e Senha
class LoginSenha : public Autenticacao {
    string login;
    string senha;

public:
    void coletarDados() override {
        cout << "--- Login via Usuario e Senha ---" << endl;
        cout << "Digite o login: ";
        cin >> login;
        cout << "Digite a senha: ";
        cin >> senha;
    }

    bool validarDados() override {
        // Regra: login > 4 letras e senha > 8 digitos
        return (login.length() > 4 && senha.length() > 8);
    }
};

// Forma 2: Login com o Google
class GoogleLogin : public Autenticacao {
    string email;

public:
    void coletarDados() override {
        cout << "--- Login via Google ---" << endl;
        cout << "Digite seu e-mail do Gmail: ";
        cin >> email;
    }

    bool validarDados() override {
        // Valida se termina com @gmail.com
        string sufixo = "@gmail.com";
        if (email.length() >= sufixo.length()) {
            return (email.compare(email.length() - sufixo.length(), sufixo.length(), sufixo) == 0);
        }
        return false;
    }
};

// Forma 3: QR Code
class QRCodeLogin : public Autenticacao {
    string codigo;

public:
    void coletarDados() override {
        cout << "--- Login via QR Code ---" << endl;
        cout << "Escaneie (digite) o código do QR Code: ";
        cin >> codigo;
    }

    bool validarDados() override {
        // So aceita se for o codigo especifico
        return (codigo == "QR12345");
    }
};

// Programa Principal
int main() {
    int opcao;
    Autenticacao* sistema = nullptr; // Ponteiro para usar polimorfismo

    cout << "Como deseja logar no Discord?" << endl;
    cout << "1 - Login/Senha\n2 - Google\n3 - QR Code" << endl;
    cout << "Opcao: ";
    cin >> opcao;

    // Escolhe qual classe instanciar baseada na opcao
    switch (opcao) {
        case 1:
            sistema = new LoginSenha();
            break;
        case 2:
            sistema = new GoogleLogin();
            break;
        case 3:
            sistema = new QRCodeLogin();
            break;
        default:
            cout << "Opcao invalida!" << endl;
            return 0;
    }

    // Executa o processo de autenticar de forma polimorfica
    if (sistema != nullptr) {
        sistema->autenticar();
        delete sistema; // Limpa a memoria
    }

    return 0;
}
```
